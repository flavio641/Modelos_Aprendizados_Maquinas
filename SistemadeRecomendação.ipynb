```python
import numpy as np 
import pandas as pd
from sklearn.neighbors import NearestNeighbors
```

```python
ratings = pd.read_csv('/content/BX-Book-Ratings.csv', sep= ';', encoding= 'latin-1',error_bad_lines= False)
ratings
```

```python
books = pd.read_csv('/content/BX-Books.csv', sep= ';', encoding= 'latin-1',error_bad_lines= False)
books
```

```python
users = pd.read_csv('/content/BX-Users.csv', sep= ';', encoding= 'latin-1',error_bad_lines= False)
users
```

```python
books = books[['ISBN', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Book-Title']]
books.rename(columns= {'Book-Title': 'title', 'Book-Author': 'autor','Year-Of-Publication':'year','Publisher': 'publicher'}, inplace= True)
users.rename(columns= {'User-ID': 'User_id', 'Location': 'location', 'Age': 'age'}, inplace= True)
ratings.rename(columns= {'User-ID': 'user_id','Book-Rating':'rating'}, inplace= True)
```

```python
books
```

```python
# Quantidade de Ratingns por Usuário
ratings['user_id'].value_counts()
```

```python
x = ratings['user_id'].value_counts() > 200
```

```python
y = x[x].index
print(y.shape)
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
# Juntando as Tabelas com (merge)
ratings_whith_books = ratings.merge(books, on= 'ISBN')
ratings_whith_books
```

```python
#Quantidade ratings dos Livros
number_ratings = ratings_whith_books.groupby('title')['rating'].count().reset_index()
```

```python
number_ratings
```

```python
number_ratings.rename(columns= {'rating': 'number_of_rating'}, inplace= True)
number_ratings
```

```python
final_rating = ratings_whith_books.merge(number_ratings, on = 'title')
final_rating
```

```python
final_rating = final_rating[final_rating['number_of_rating']> 50]
final_rating.shape
```

```python
# Vamos descartar valores duplicados
final_rating.drop_duplicates(['user_id', 'title'], inplace= True)
final_rating.shape
```

```python
# Transposição das linhas(user_id) em colunas 
book_pivot = final_rating.pivot_table(columns= 'user_id', index= 'title', values = 'rating')
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace= True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparce = csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
model.fit(book_sparce)
```

```python
distancias, sugestoes = model.kneighbors(book_pivot.iloc[258, :].values.reshape(1, -1))
```

```python
for i in range(len(sugestoes)):
  print(book_pivot.index[sugestoes[i]])
```
